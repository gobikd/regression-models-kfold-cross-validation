```python
import pandas as pd# import=load library/module so can be used in code. pandas=Excel+SQL and moaths tools in Python. 
import numpy as np
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression #CLASS LinearRegression inported from PACKAGE.MODULE: sklearn.linear_model
regressor=LinearRegression()#creates instant from class. Still not learned,empty
```

```python
dataset=pd.read_csv("../data/Salary_Data.csv")
```

```python
dataset #dataset now is pandas's dataframe. This line displays full data
```

```python
independent=dataset[["YearsExperience"]]# Those in bracket same as in CVS and dataset 1st row displayed. Input X
#note double bracket needed here since table/dataframe expexted by sklean (another module/function later)
independent# List down datafam 
```

```python
dependent=dataset[["Salary"]] #Now to Y dataset, Output variables Y. Note double brackets. Passes dataframe Salary to  dependent
dependent# listing
```

```python
print("="*60)
print("independent shape:", independent.shape)
print("dependent shape:", dependent.shape)
print("="*60)
```

```python
# Since K_Fold will do import train_test_split -disabled#
#from sklearn.model_selection import train_test_split 
#X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
#-----------K-Fold--------------#
kfold = KFold(
    n_splits=10,
    shuffle=True,
    random_state=0
)

scores = cross_val_score(
    estimator=regressor, #!! Regressor assigned
    X=independent,       #!! Data X and Y assigned
    y=dependent,
    cv=kfold,            #!! 
    scoring="r2"
)
#-----------K-Fold--------------#
print("="*60)
print("R2 Score for each fold:", scores)
print("Mean R2:", scores.mean())
print("Standard Deviation:", scores.std())
print("="*60)
```

```python
# Train final model on the complete dataset
regressor.fit(independent, dependent)
# Note the model was trained with overall dataset-no learning
```

```python
weight=regressor.coef_# slope coefficient(W), yWx+b, taken from fc regressor.coef_ assigned to weight
weight# display result stored in weight which is equal to regressor.coef_
# Note, Compare with trained result weight=9360.26128619
```

```python
bias=regressor.intercept_ #spelling for bais was changed to bias. Actually value stored in regressor.intercept passed to bias.
bias# display value stored in bias equal to regressor.intercept_ a fc defined within LinearRegression(),CLASS,
# Note, Compare with trained result bias=26777.3913412
```

```python
# y_pred=regressor.predict(X_test) will not work
# from sklearn.metrics import r2_score
# r_score=r2_score(y_test,y_pred)
```

```python
# KFold produce r2 result for each fold only which can be visualised
# Now compre r2=0.9740993407213511 from regressor.fit(X_train,y_train) after train_test_split method againt
# 10 fold of KFold. 
import matplotlib.pyplot as plt
plt.bar(range(1, 11), scores)
plt.xlabel("Fold")
plt.ylabel("R²")
plt.title("R² Score for Each Fold")
plt.show()
```

```python
plt.plot(range(1,11), scores, marker='o')
plt.xlabel("Fold")
plt.ylabel("R²")
plt.title("R² Across 10 Folds")
plt.grid(True)
plt.show()
```

```python
plt.boxplot(scores)
plt.ylabel("R²")
plt.title("Distribution of R² Scores")
plt.show()
```
